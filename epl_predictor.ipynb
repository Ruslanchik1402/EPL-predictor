1. Считываем информацию с csv-файлов, отбираем только нужные для прогнозов данные, объединяем все датафреймы в один и исключаем команды, которые не принимают участия в текущем сезоне.

```python
import pandas as pd



files=["season-1011_csv.csv","season-1112_csv.csv","season-1213_csv.csv","season-1314_csv.csv","season-1415_csv.csv","season-1516_csv.csv","season-1617_csv.csv","season-1718_csv.csv","season-1819_csv.csv"]

l=[]

for f in files:
    season_data=pd.read_csv(f)
    l.append(season_data)
    
    
season_data_list=l


columns_names=['Div', 'Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'HTR', 'Referee', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR']
needed_columns=set(columns_names)

for a in l:
    all_columns=set(list(a))
    columns_to_drop=list(all_columns.difference(needed_columns))
    a.drop(columns_to_drop, axis = 1, inplace = True)
    
final_data=pd.concat(l,ignore_index=True)

actual_teams=["Liverpool", "Man City", "Chelsea", "Leicester", "Man United", "Tottenham", "Wolves", "Burnley", "Arsenal", "Southampton", "Everton", "Newcastle", "Crystal Palace", "Brighton", "West Ham", "Watford", "Aston Villa", "Bournemouth", "Norwich"]
final_data=final_data.loc[final_data['HomeTeam'].isin(actual_teams) & final_data['AwayTeam'].isin(actual_teams)]
final_data["Season"]='-'
```

2. Прописываем функцию, которой на вход подается дата проведения матча, а на возврат будет string с сезоном. Try/except использую здесь для корректного считывания даты, потому что в наших csv-файлах способ указания даты разнится.

```python
from datetime import datetime


def ret_season(date):
    try:
        date=datetime.strptime(date, '%d/%m/%y')
    except:
        date=datetime.strptime(date, '%d/%m/%Y')
    date = date.strftime('%d/%m/%y')
    res=" "
    date1011_start = '1.07.2010'
    date1011_finish = '1.07.2011'
    date1112_start = '2.07.2011'
    date1112_finish = '1.07.2012'
    date1213_start = '2.07.2012'
    date1213_finish = '1.07.2013'
    date1314_start = '2.07.2013'
    date1314_finish = '1.07.2014'
    date1415_start = '2.07.2014'
    date1415_finish = '1.07.2015'
    date1516_start = '2.07.2015'
    date1516_finish = '1.07.2016'
    date1617_start = '2.07.2016'
    date1617_finish = '1.07.2017'
    date1718_start = '2.07.2017'
    date1718_finish = '1.07.2018'
    date1819_start = '2.07.2018'
    date1819_finish = '1.07.2019'
    if (datetime.strptime(date, '%d/%m/%y')>=datetime.strptime(date1011_start, '%d.%m.%Y')) and (datetime.strptime(date, '%d/%m/%y')<=datetime.strptime(date1011_finish, '%d.%m.%Y')):
        res="10/11"
    elif datetime.strptime(date, '%d/%m/%y')>=datetime.strptime(date1112_start, '%d.%m.%Y') and datetime.strptime(date, '%d/%m/%y')<=datetime.strptime(date1112_finish, '%d.%m.%Y'):
        res="11/12"
    elif datetime.strptime(date, '%d/%m/%y')>=datetime.strptime(date1213_start, '%d.%m.%Y') and datetime.strptime(date, '%d/%m/%y')<=datetime.strptime(date1213_finish, '%d.%m.%Y'):
        res="12/13"
    elif datetime.strptime(date, '%d/%m/%y')>=datetime.strptime(date1314_start, '%d.%m.%Y') and datetime.strptime(date, '%d/%m/%y')<=datetime.strptime(date1314_finish, '%d.%m.%Y'):
        res="13/14"
    elif datetime.strptime(date, '%d/%m/%y')>=datetime.strptime(date1415_start, '%d.%m.%Y') and datetime.strptime(date, '%d/%m/%y')<=datetime.strptime(date1415_finish, '%d.%m.%Y'):
        res="14/15"
    elif datetime.strptime(date, '%d/%m/%y')>=datetime.strptime(date1516_start, '%d.%m.%Y') and datetime.strptime(date, '%d/%m/%y')<=datetime.strptime(date1516_finish, '%d.%m.%Y'):
        res="15/16"
    elif datetime.strptime(date, '%d/%m/%y')>=datetime.strptime(date1617_start, '%d.%m.%Y') and datetime.strptime(date, '%d/%m/%y')<=datetime.strptime(date1617_finish, '%d.%m.%Y'):
        res="16/17"
    elif datetime.strptime(date, '%d/%m/%y')>=datetime.strptime(date1718_start, '%d.%m.%Y') and datetime.strptime(date, '%d/%m/%y')<=datetime.strptime(date1718_finish, '%d.%m.%Y'):
        res="17/18"
    elif datetime.strptime(date, '%d/%m/%y')>=datetime.strptime(date1819_start, '%d.%m.%Y') and datetime.strptime(date, '%d/%m/%y')<=datetime.strptime(date1819_finish, '%d.%m.%Y'):
        res="18/19"
    return res
    


a=list(final_data['Date'])
seasons=[]
for x in a:
    seasons.append(ret_season(x))
    
final_data['Season']=seasons
```

3. Функция, возвращающая список со статистикой команды за определенный сезон

```python
def SeasonStats(dataframe,team, season,dec): ####dec - отображать статистику с названиями показателей или нет 1=да, 0=нет
    data=dataframe
    goalScored = 0 
    goalAllowed = 0 
    
    HTGoalsScored = 0
    HTGoalsAllowed = 0
    
    gamesWon = 0 
    gamesDrawn = 0 
    gamesLost = 0 
    
    HTWin = 0
    HTDraw = 0
    HTLost = 0
    
    totalScore = 0 
    
    matches = 0 
    
    
    ShotsMade = 0
    ShotsAllowed = 0
    ShotsOnTargetMade = 0
    ShotsOnTargetAllowed = 0
    
    
    FoulsMade = 0
    FoulsReceived = 0
    
    
    CornersMade = 0
    CornersAllowed = 0
    
    YellowCardEarned = 0
    YellowCardAllowed = 0
    
    RedCardEarned = 0
    RedCardAllowed = 0

    for index, row in data.iterrows():
        if ((row['Season'] == season) & (row['HomeTeam']==team))|((row['Season'] == season) & (row['AwayTeam']==team)):
            matches+=1
            
            if (row['HomeTeam']==team):
                goalScored += row['FTHG']
                goalAllowed += row['FTAG']
                HTGoalsScored +=row['HTHG']
                HTGoalsAllowed +=row['HTAG']
                ShotsMade += row['HS']
                ShotsAllowed += row['AS']
                ShotsOnTargetMade += row['HST']
                ShotsOnTargetAllowed += row['AST']
                FoulsMade += row['HF']
                FoulsReceived += row['AF']
                CornersMade += row['HC']
                CornersAllowed += row['AC']
                YellowCardEarned += row['HY']
                YellowCardAllowed += row['AY']
                RedCardEarned += row['HR']
                RedCardAllowed += row['AR']
                
            
                if (row['FTR']=='H'):
                    totalScore += 3
                    gamesWon += 1
                elif (row['FTR']=='D'):
                    totalScore += 1
                    gamesDrawn += 1
                elif (row['FTR']=='A'):
                    totalScore += 0
                    gamesLost += 1
                if (row['HTR']=='H'):
                    HTWin += 1
                elif (row['HTR']=='D'):
                    HTDraw += 1
                elif (row['HTR']=='A'):
                    HTLost += 1
            elif (row['AwayTeam']==team):
                goalAllowed += row['FTHG']
                goalScored += row['FTAG']
                HTGoalsAllowed +=row['HTHG']
                HTGoalsScored +=row['HTAG']
                ShotsAllowed += row['HS']
                ShotsMade += row['AS']
                ShotsOnTargetAllowed += row['HST']
                ShotsOnTargetMade += row['AST'] 
                FoulsReceived += row['HF']
                FoulsMade += row['AF']
                CornersAllowed += row['HC']
                CornersMade += row['AC']
                YellowCardAllowed += row['HY']
                YellowCardEarned += row['AY']
                RedCardAllowed += row['HR']
                RedCardEarned += row['AR']
                if (row['FTR']=='A'):
                    totalScore += 3
                    gamesWon += 1
                elif (row['FTR']=='D'):
                    totalScore += 1
                    gamesDrawn += 1
                elif (row['FTR']=='H'):
                    totalScore += 0
                    gamesLost += 1
                if (row['HTR']=='A'):
                    HTWin += 1
                elif (row['HTR']=='D'):
                    HTDraw += 1
                elif (row['HTR']=='H'):
                    HTLost += 1
    stats_names=['Games played','Points','Won','Drawn','Lost','Half Time won',
                '1st half won','1st half drawn','1st half lost',
                'Goals scored','Goals conceded',
                '1st half goals scored','1st half goals conceded',
                'Shots made','Shots on target made',
                'Shots allowed','Shots on target allowed',
                'Team fouls','Opposite fouls',
                'Team corners','Opposite corners',
                'Team yellow cards','Opposite yellow cards',
                'Team red cards','Oppostie red cards']
    
    res_stats=[matches, totalScore, 
            gamesWon, gamesDrawn, gamesLost,
            HTWin, HTDraw, HTLost,
            goalScored, goalAllowed, 
            HTGoalsScored, HTGoalsAllowed, 
            ShotsMade, ShotsOnTargetMade, ShotsAllowed, ShotsOnTargetAllowed, 
            FoulsMade, FoulsReceived,
            CornersMade, CornersAllowed,
            YellowCardEarned, YellowCardAllowed,
            RedCardEarned, RedCardAllowed]
    if dec==1:
        for sn,s in zip(stats_names,res_stats):
            print(sn,s)
    
    return res_stats
            
            
```

4. Функция, составляющая словарь со статистиками всех команд за определенный сезон.

```python
from collections import defaultdict

def AllTeamsSeasonStats(dataframe,season):
    annual = defaultdict(list)
    TeamsList=dataframe['HomeTeam'].unique()
    for team in TeamsList:
        team_vector = SeasonStats(dataframe,team, season,0)
        annual[team] = team_vector
    return annual
```

5. Функция для создания обучающей выборки из имеющихся данных.

```python
import numpy as np

def GetTrainingData(dataframe,seasons):
    totalNumGames = 0
    for season in seasons:
        annual = dataframe[dataframe['Season'] == season]
        totalNumGames += len(annual.index)
    numFeatures = len(SeasonStats(dataframe,'Arsenal', '15/16',0)) 
    xTrain = np.zeros(( totalNumGames, numFeatures))
    yTrain = np.zeros(( totalNumGames ))
    indexCounter = 0
    for season in seasons:
        team_vectors = AllTeamsSeasonStats(dataframe,season)
        annual = dataframe[dataframe['Season'] == season]
        numGamesInYear = len(annual.index)
        xTrainAnnual = np.zeros(( numGamesInYear, numFeatures))
        yTrainAnnual = np.zeros(( numGamesInYear ))
        counter = 0
        for index, row in annual.iterrows():
            team = row['HomeTeam']
            t_vector = team_vectors[team]
            rivals = row['AwayTeam']
            r_vector = team_vectors[rivals]
           
            diff = [a - b for a, b in zip(t_vector, r_vector)]
            
            if len(diff) != 0:
                xTrainAnnual[counter] = diff
            if row['FTR'] == 'H':
                yTrainAnnual[counter] = 1
            else: 
                yTrainAnnual[counter] = 0
            counter += 1   
        xTrain[indexCounter:numGamesInYear+indexCounter] = xTrainAnnual
        yTrain[indexCounter:numGamesInYear+indexCounter] = yTrainAnnual
        indexCounter += numGamesInYear
    return xTrain, yTrain


xTrain, yTrain = GetTrainingData(final_data,['13/14','14/15','15/16','16/17','17/18','18/19'])
```

6. Обучение модели.

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(xTrain, yTrain)
```

7. Функция для получения прогноза на матч.

```python
def GetGamePrediction(team1_vector, team2_vector,model):
    diff = [[a - b for a, b in zip(team1_vector, team2_vector)]]
    predictions = model.predict(diff)
    return predictions
```
